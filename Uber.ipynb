```python
#import libraries
import numpy as np
import pandas as pd
```

```python
#import dataset
df=pd.read_csv('/content/uber_ncr_ride_bookings.csv')
```

```python
#checking for data
df.head()
```

```python
#check null values in dataset
df.isnull().sum()
```

```python
#checing data types of columns
df.info()
```

```python
#fill missing values
categorical_cols = ['Reason for cancelling by Customer', 'Driver Cancellation Reason', 'Incomplete Rides Reason']
for col in categorical_cols:
    df[col] = df[col].fillna('Unknown')
```

```python
#fill missing values
numerical_cols = ['Avg VTAT', 'Avg CTAT', 'Booking Value', 'Ride Distance']
for col in numerical_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
#fill 0  in missing values
status_cols = ['Cancelled Rides by Customer', 'Cancelled Rides by Driver', 'Incomplete Rides']
for col in status_cols:
    df[col] = df[col].fillna(0)
```

```python
#fill 0  in missing values
rating_cols = ['Driver Ratings']
for col in rating_cols:
  df[col] = df[col].fillna(0)
```

```python
#fill 0  in missing values
cus_rating_cols = ['Customer Rating']
for col in cus_rating_cols:
  df[col] = df[col].fillna(0)
```

```python
#fill 'Unknown' in 'Payment Method' column
df['Payment Method'] = df['Payment Method'].fillna('Unknown')
```

```python
#cheking null values after fill values
df.isnull().sum()
```

```python
#checking for duplicate values
df.duplicated().sum()
```

```python
#replace data type of columns in category from object
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].astype('category')
```

```python
#checking for data type
df.info()
```

```python
#replace all spaces in column names in underscore
df.columns = df.columns.str.replace(' ', '_')
```

```python
#checking for column's names
df.columns
```

```python
#checking for nunique values
df.nunique()
```

```python
#use statistics of data set
df.describe()
```

```python
#data is cleaned & save the data set with changes
df.to_csv('uber_cleaned.csv', index=False)
```
